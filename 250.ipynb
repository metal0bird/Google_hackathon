```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
le=preprocessing.LabelEncoder()
import os
```

```python
train_data=pd.DataFrame()
for dirname, _, filenames in os.walk('train_raw'):
    for filename in filenames:
        fold=pd.read_csv(os.path.join(dirname, filename))
        train_data=pd.concat([train_data,fold])

test_data=pd.read_csv('test_raw.csv/test.csv')
```

```python
train_drop = train_data.dropna()
test_drop = train_data.dropna()
train_drop.shape
```

```python
train_drop['malware_status']=train_data['malware_status']
```

```python
train_drop_after_delete = train_drop.dropna()
train_drop_after_delete.shape
```

```python
#imputing the missing values with mode 
train_drop_after_delete_mode = train_drop.fillna(train_drop.mode().iloc[0])
```

```python
del train_drop_after_delete_mode['machine_id']
```

```python
#applying data encoding 
train_drop_mode_encoded=train_drop_after_delete_mode.apply(le.fit_transform)
train_drop_mode_encoded
```
